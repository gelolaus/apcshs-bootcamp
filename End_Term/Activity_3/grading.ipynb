```python
# Created by Angelo John Benedict Laus
# For Asia Pacific College - SHS SoCIT Bootcamp A.Y. 2022-2023
# Import getpass for password
import getpass

def main():
    password = "jupyter"
    attempts = 0
    
    print("Created by Angelo Laus \nFor Asia Pacific College - SHS SoCIT Bootcamp A.Y. 2022-2023 \n")

# Ask for the user's password
    while attempts < 3:
        user_input = getpass.getpass("Enter password: ")
        
        if user_input == password:
            grading()
            break
        else:
            attempts += 1
            print("Incorrect password!")
            if attempts == 3:
                print("You have been locked out!")

def grading():
    print("\nWelcome! to the Grading System")
    print("\n")

# Get the input from the user
    quiz1 = float(input("Enter Quiz 1 score: "))
    while quiz1 < 0 or quiz1 > 100:
        print("No input can be less than 0 or higher than 100")
        quiz1 = float(input("Enter Quiz 1 score: "))

    quiz2 = float(input("Enter Quiz 2 score: "))
    while quiz2 < 0 or quiz2 > 100:
        print("No input can be less than 0 or higher than 100")
        quiz2 = float(input("Enter Quiz 2 score: "))
    
    quiz3 = float(input("Enter Quiz 3 score: "))
    while quiz3 < 0 or quiz3 > 100:
        print("No input can be less than 0 or higher than 100")
        quiz3 = float(input("Enter Quiz 3 score: "))

    project1 = float(input("Enter Project 1 score: "))
    while project1 < 0 or project1 > 100:
        print("No input can be less than 0 or higher than 100")
        project1 = float(input("Enter Project 1 score: "))

    project2 = float(input("Enter Project 2 score: "))
    while project2 < 0 or project2 > 100:
        print("No input can be less than 0 or higher than 100")
        project2 = float(input("Enter Project 2 score: "))

    project3 = float(input("Enter Project 3 score: "))
    while project3 < 0 or project3 > 100:
        print("No input can be less than 0 or higher than 100")
        project3 = float(input("Enter Project 3 score: "))

    assignment1 = float(input("Enter Assignment 1 score: "))
    while assignment1 < 0 or assignment1 > 100:
        print("No input can be less than 0 or higher than 100")
        assignment1 = float(input("Enter Assignment 1 score: "))

    assignment2 = float(input("Enter Assignment 2 score: "))
    while assignment2 < 0 or assignment2 > 100:
        print("No input can be less than 0 or higher than 100")
        assignment2 = float(input("Enter Assignment 2 score: "))

    final_exam = float(input("Enter Final Exam score: "))
    while final_exam < 0 or final_exam > 100:
        print("No input can be less than 0 or higher than 100")
        final_exam = float(input("Enter Final Exam score: "))

# Compute for the final grade
    quiz_average = (quiz1 + quiz2 + quiz3) / 3
    project_average = (project1 + project2 + project3) / 3
    assignment_average = (assignment1 + assignment2) / 2
    final_exam_average = final_exam
    final_grade = (quiz_average * 0.3) + (project_average * 0.2) + (assignment_average * 0.1) + (final_exam_average * 0.4)

# Numerical Equivalent and Remarks
    if final_grade >= 95:
        numerical_equivalent = 4.0
    elif final_grade >= 91:
        numerical_equivalent = 3.5
    elif final_grade >= 87:
        numerical_equivalent = 3.0
    elif final_grade >= 83:
        numerical_equivalent = 2.5
    elif final_grade >= 79:
        numerical_equivalent = 2.0
    elif final_grade >= 75:
        numerical_equivalent = 1.5
    elif final_grade >= 70:
        numerical_equivalent = 1.0
    else:
        numerical_equivalent = "R"

    if numerical_equivalent >= 2.0 or numerical_equivalent != "R":
        remarks = "Passed"
    else:
        remarks = "Failed"

# Display the final results
    print("\n")
    print("Quiz Average:", format(quiz_average, ".2f"))
    print("Project Average:", format(project_average, ".2f"))
    print("Assignment Average:", format(assignment_average, ".2f"))
    print("Final Exam Average:", format(final_exam_average, ".2f"))
    print("\n")
    print("Final Grade:", format(final_grade, ".2f"))
    print("Numerical Equivalent:", numerical_equivalent)
    print("Remarks:", remarks)
    end()
    


def end():
    print("\nDo you want to try again?")
    print("Type \"1\" if Yes")
    print("Type \"2\" No")
    # Ask the user if they want to try again
    user_input = int(input("Enter choice: "))
    if user_input == 1:
        grading()
    elif user_input == 2:
        print("\nThank you for using the Grading System!")
    else:
        print("Invalid input!")
        end()

main()
```
